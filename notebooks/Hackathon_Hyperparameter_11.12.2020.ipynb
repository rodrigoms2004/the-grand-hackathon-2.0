```python
# Create a dataframe for storing model results
res = pd.DataFrame({'Model':[],
                    'R2 Train': [],
                    'R2 Test': [],
                    'MAPE Train': [],
                    'MAPE Test': [],
                    'MAE Train': [],
                    'MAE Test': [],
                   })

res
```

```python
# Creating a function for trying multiple models at once
def try_models(models_to_try):
    for name, model in models_to_try:
        model.fit(X_train, y_train)
        res.loc[len(res)] = [name, 
                             r2_score(y_train, model.predict(X_train)),                              # R2 on Train data
                             r2_score(y_test, model.predict(X_test)),                                # R2 on Test data                                     
                             np.mean(np.abs((y_train - model.predict(X_train)) / y_train)) * 100,    # Mean Absolute Percentage Error (MAPE) on Train data
                             np.mean(np.abs((y_test - model.predict(X_test)) / y_test)) * 100 ,      # Mean Absolute Percentage Error (MAPE) on Test data
                             np.mean(np.abs(y_train - model.predict(X_train))),    # Mean Absolute Percentage Error (MAPE) on Train data
                             np.mean(np.abs(y_test - model.predict(X_test)))]   
    return res

# Creating a function for visualizing refressor results
def visRegressorResults(model_w_parameters):
    visualizer = ResidualsPlot(model_w_parameters)
    visualizer.fit(X_train, y_train)  
    visualizer.score(X_test, y_test) 
    visualizer.show()

# Creating a function for showing feature importance
def show_feature_importance(model, label):
    
    imp = dict(zip(X_train.columns, model.feature_importances_))
    imp = pd.DataFrame.from_dict(imp, orient='index').reset_index()
    imp = imp.rename(columns={"index": "Feature",
                                    0: "Importance"})
    imp['Importance'] = round(imp['Importance']*100, 1)
    imp = imp.sort_values('Importance')

    fig = px.bar(imp, 
             y = 'Feature', 
             x = 'Importance', 
             orientation ='h', 
             text = 'Importance', 
             title = label + ': Feature Importance', 
             width = 1300, 
             height = 500)

    fig.update_traces(texttemplate='%{text:.1f}', textposition = 'outside')

    fig.show()

    
# Creating a function for evaluationg a model without training
def evaluate_model(model):
    
    r2_train = r2_score(y_train, model.predict(X_train))                           
    r2_test = r2_score(y_test, model.predict(X_test))                                                             
    mape_train = np.mean(np.abs((y_train - model.predict(X_train)) / y_train)) * 100
    mape_test = np.mean(np.abs((y_test - model.predict(X_test)) / y_test)) * 100    
    mae_train = np.mean(np.abs(y_train - model.predict(X_train)) ) 
    mae_test = np.mean(np.abs(y_test - model.predict(X_test))) 
        
    return r2_train, r2_test, mape_train, mape_test,mae_train,mae_test
```

```python
#Bayesian optimization
def bayesian_optimization(dataset, function, parameters):
   X_train, y_train, X_test, y_test = dataset
   n_iterations = 5
   gp_params = {"alpha": 1e-4}

   BO = BayesianOptimization(function, parameters)
   BO.maximize(n_iter=n_iterations, **gp_params)

   return BO.max
```

```python
def rfc_optimization(cv_splits):
    def function(n_estimators, max_depth, min_samples_split):
        return cross_val_score(
               RandomForestRegressor(
                   n_estimators=int(max(n_estimators,0)),                                                               
                   max_depth=int(max(max_depth,1)),
                   min_samples_split=int(max(min_samples_split,2)), 
                   n_jobs=-1, 
                   random_state=42),  
               X=X_train, 
               y=y_train, 
               cv=cv_splits,
               scoring="neg_mean_absolute_error",
               n_jobs=-1).mean()

    parameters = {"n_estimators": (10, 1000),
                  "max_depth": (1, 150),
                  "min_samples_split": (2, 10)
                  }
    
    print('INNerFunction is called ')
    return function, parameters
```

```python
#Train model
def train(X_train, y_train, X_test, y_test, function, parameters):
    dataset = (X_train, y_train, X_test, y_test)
    cv_splits = 4
    
    best_solution = bayesian_optimization(dataset, function, parameters)      
    params = best_solution["params"]

    model = RandomForestClassifier(
             n_estimators=int(max(params["n_estimators"], 0)),
             max_depth=int(max(params["max_depth"], 1)),
             min_samples_split=int(max(params["min_samples_split"], 2)), 
             n_jobs=-1, 
             random_state=42)

    model.fit(X_train, y_train)
    
    return model
```

```python
# %%time
rf = RandomForestRegressor(random_state=None)
# Define a cross-validation generator
k =  KFold(random_state=7, n_splits=5)

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 5 , stop = 1000, num = 200)]   # returns evenly spaced 25 numbers
# Number of features to consider at every split
max_features = ['log2', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 10, num = 5)]  # returns evenly spaced numbers can be changed to any
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2,3,4,5,6,7,8,9,10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 3, 4,5]
# Method of selecting samples for training each tree
#learning_rate = [float(x) for x in np.linspace(0.001, 1, num = 100)]

params = {
    'bootstrap': [True,False],
    'max_depth': max_depth,
    'max_features': ['sqrt','log2'],
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'n_estimators': n_estimators
}    



rf_cv = RandomizedSearchCV(estimator = rf, param_distributions = params, cv = k, verbose = 1, n_jobs = 4)

rf_cv.fit(X_train, y_train)

# Check best parameters
rf_cv.best_params_
```

```python
gb = GradientBoostingRegressor(random_state=7)

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 5 , stop = 1000, num = 200)]   # returns evenly spaced 25 numbers
# Number of features to consider at every split
max_features = ['log2', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 10, num = 5)]  # returns evenly spaced numbers can be changed to any
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2,3,4,5,6,7,8,9,10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 3, 4]
# Method of selecting samples for training each tree
learning_rate = [float(x) for x in np.linspace(0.001, 1, num = 100)]

# Create the random grid
params_r = {'n_estimators': n_estimators,
            'max_features': max_features,
            'max_depth': max_depth,
            'min_samples_split': min_samples_split,
            'min_samples_leaf': min_samples_leaf,
            'learning_rate':learning_rate}

# Use the random grid to search for best hyperparameters

# Random search of parameters using cross validation
gb_cv = RandomizedSearchCV(estimator=gb, param_distributions=params_r,cv = k, random_state=7, verbose = 1, n_jobs = 4, n_iter = 100)

# Fit the random search model
gb_cv.fit(X_train,y_train)

# Check best parameters
gb_cv.best_params_
```
